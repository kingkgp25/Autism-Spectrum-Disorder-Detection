### Data-3

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
data_3 = pd.read_csv('../DataSet/Screening/Autism Child/csv_result-Autism-Child-Data.csv')
```

```python
data_3
```

```python
data_3['relation'].unique()
```

```python
data_3['relation'].replace({'self':'Self', 'Parent':'family member', 'Relative':'Others'}, inplace = True)
```

```python
data_3['relation'].unique()
```

```python
display(data_3)
```

#### Converting binary categorial variable to numerical variables

```python
data_3['gender'].replace(['f','m'], [0,1] ,inplace=True)
data_3['jundice'].replace(['no','yes'], [0,1] ,inplace=True)
data_3['austim'].replace(['no','yes'], [0,1] ,inplace=True)
data_3['used_app_before'].replace(['no','yes'], [0,1] ,inplace=True)
data_3['Class/ASD'].replace(['NO','YES'], [0,1] ,inplace=True)
```

```python
data_3
```

```python
print("Counting ? values\n")
print(data_3.isin(['?']).sum(axis=0))
```

#### Feature Scaling

```python
enc = LabelEncoder()
categ = ['ethnicity','contry_of_res','relation']
data_3[categ] = data_3[categ].apply(enc.fit_transform)
```

```python
data_3
```

#### Filling missing values

```python
data_3['ethnicity'] = data_3['ethnicity'].replace(0,np.nan)
data_3['relation'] = data_3['relation'].replace(0,np.nan)
data_3['age'] = data_3['age'].replace('?',np.nan)
```

```python
print(data_3.isnull().any().any())
print(data_3.isnull().sum().sum())
```

```python
data_3.columns[data_3.isna().any()].tolist()
```

```python
data_3['age'].fillna(data_3['age'].mean(), inplace = True)
```

```python
print(data_3.isnull().any().any())
print(data_3.isnull().sum().sum())
```
