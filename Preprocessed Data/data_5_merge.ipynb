#### Converting binary categorial variable to numerical variables

```python
data_5['gender'].replace(['f','m'], [0,1] ,inplace=True)
data_5['jundice'].replace(['no','yes'], [0,1] ,inplace=True)
data_5['austim'].replace(['no','yes'], [0,1] ,inplace=True)
data_5['used_app_before'].replace(['no','yes'], [0,1] ,inplace=True)
data_5['Class/ASD'].replace(['NO','YES'], [0,1] ,inplace=True)
```
