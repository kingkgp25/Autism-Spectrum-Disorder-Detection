### Data-2

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
data_2 = pd.read_csv('../DataSet/Screening/Autism adults/csv_result-Autism-Adult-Data.csv')
```

```python
data_2
```

```python
data_2.drop('id', axis=1, inplace=True)
```

```python
display(data_2)
```

#### Converting binary categorial variable to numerical variables

```python
data_2['gender'].replace(['f','m'], [0,1] ,inplace=True)
data_2['jundice'].replace(['no','yes'], [0,1] ,inplace=True)
data_2['austim'].replace(['no','yes'], [0,1] ,inplace=True)
data_2['used_app_before'].replace(['no','yes'], [0,1] ,inplace=True)
data_2['Class/ASD'].replace(['NO','YES'], [0,1] ,inplace=True)
```

```python
data_2['age'].unique()
```

```python
data_2['contry_of_res'].unique()
```

```python
print("Counting ? values\n")
print(data_2.isin(['?']).sum(axis=0))
```

#### Feature Scaling

```python
enc = LabelEncoder()
categ = ['ethnicity','contry_of_res','relation']
data_2[categ] = data_2[categ].apply(enc.fit_transform)
```

```python
data_2
```

#### Filling missing values

```python
data_2["age"].mode()
```

```python
data_2['ethnicity'] = data_2['ethnicity'].replace('?',np.nan)
data_2['relation'] = data_2['relation'].replace('?',np.nan)
data_2['age'] = data_2['age'].replace('?',0)
```

```python
print(data_2.isnull().any().any())
print(data_2.isnull().sum().sum())
```

```python
data_2['relation'].unique()
```

```python
data_2['age'].unique()
```

```python
data_2['contry_of_res'].unique()
```

```python
data_2.columns[data_2.isna().any()].tolist()
```

```python
data_2['ethnicity'].fillna(data_2['ethnicity'].mean(), inplace = True)
data_2['relation'].fillna(data_2['relation'].mean(), inplace = True)
data_2['age'].fillna(data_2["age"].mode(), inplace = True)
```

```python
print(data_2.isnull().any().any())
print(data_2.isnull().sum().sum())
```
