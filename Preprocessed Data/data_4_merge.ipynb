#### Converting binary categorial variable to numerical variables

```python
data_4['Sex'].replace(['f','m'], [0,1] ,inplace=True)
data_4['Jaundice'].replace(['no','yes'], [0,1] ,inplace=True)
data_4['Family_mem_with_ASD'].replace(['no','yes'], [0,1] ,inplace=True)
data_4['Class/ASD Traits '].replace(['No','Yes'], [0,1] ,inplace=True)
```

```python

print("Counting ? values\n")
print(data_4.isin(['?']).sum(axis=0))
```

#### Feature Scaling

```python
enc = LabelEncoder()
categ = ['Ethnicity','Who completed the test']
data_4[categ] = data_4[categ].apply(enc.fit_transform)
```
