### Data-2.2

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
data_2_2 = pd.read_csv('D:\COLLEGE\CODING\Github Repos\DeepDet\DataSet\Screening/Autism adults/autism_screening_2.csv')
```

```python
data_2_2
```

#### Converting binary categorial variable to numerical variables

```python
data_2_2['gender'].replace(['f','m'], [0,1] ,inplace=True)
data_2_2['jundice'].replace(['no','yes'], [0,1] ,inplace=True)
data_2_2['austim'].replace(['no','yes'], [0,1] ,inplace=True)
data_2_2['used_app_before'].replace(['no','yes'], [0,1] ,inplace=True)
data_2_2['Class/ASD'].replace(['NO','YES'], [0,1] ,inplace=True)
```

```python
data_2_2
```

```python
print("Counting ? values\n")
print(data_2_2.isin(['?']).sum(axis=0))
```

#### Feature Scaling

```python
enc = LabelEncoder()
categ = ['ethnicity','contry_of_res','relation']
data_2_2[categ] = data_2_2[categ].apply(enc.fit_transform)
```

```python
data_2_2
```

#### Filling missing values

```python
data_2_2['ethnicity'] = data_2_2['ethnicity'].replace(0,np.nan)
data_2_2['relation'] = data_2_2['relation'].replace(0,np.nan)
data_2_2['age'] = data_2_2['age'].replace('?',np.nan)
```

```python
print(data_2_2.isnull().any().any())
print(data_2_2.isnull().sum().sum())
```

```python
data_2_2.columns[data_2_2.isna().any()].tolist()
```

```python
data_2_2['ethnicity'].fillna(data_2_2['ethnicity'].mean(), inplace = True)
data_2_2['relation'].fillna(data_2_2['relation'].mean(), inplace = True)
data_2_2['age'].fillna(0, inplace = True)
```

```python
print(data_2_2.isnull().any().any())
print(data_2_2.isnull().sum().sum())
```
